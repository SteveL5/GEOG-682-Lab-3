```python
import rasterio
from rasterio import plot
from rasterio.merge import merge
import matplotlib.pyplot as plt
```

```python
# This notebook explore two 3 band RGB Sentinel-2 scenes in a UTM projection.
#image_file = "image1.tif"

# Use Rasterio to open the images.
satdat1 = rasterio.open("image1.tif")
satdat2 = rasterio.open("image2.tif")
```

```python
plot.show(satdat1, cmap='terrain')
```

```python
plot.show(satdat2, cmap='terrain')
```

```python
#show the raster CRS
satdat1.crs
```

```python
#show the raster CRS
satdat2.crs
```

WGS 84 / UTM zone 18N

```python
#correct band order
fig, (axr, axg, axb) = plt.subplots(1,3, figsize=(21,7))
plot.show((satdat1, 1), ax=axr, cmap='Reds', title='red channel')
plot.show((satdat1, 2), ax=axg, cmap='Greens', title='green channel')
plot.show((satdat1, 3), ax=axb, cmap='Blues', title='blue channel')
```

```python
#incorrect band order
fig, (axr, axg, axb) = plt.subplots(1,3, figsize=(21,7))
plot.show((satdat1, 3), ax=axr, cmap='Reds', title='red channel')
plot.show((satdat1, 2), ax=axg, cmap='Greens', title='green channel')
plot.show((satdat1, 1), ax=axb, cmap='Blues', title='blue channel')
```
